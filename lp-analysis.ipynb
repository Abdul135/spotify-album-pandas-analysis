```python
import pickle
data = pickle.load(open('U2-lp-data.p', 'rb'))
print(data.shape)
print(data.columns)
```

```python
print(data['album'].unique())
```

```python
import pandas as pd
#data['album'] = pd.Categorical(data['album'], ['Hybrid', 'Meteora', 'Minutes',
#                                              'Thousandsuns', 'Livingthings', 'Huntingparty',
#                                              'Onemorelight'])
data['album'] = data['album'].astype('category')
```

```python
data.groupby(by='album')['album'].count()
```

## Introductory Analysis

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (24,8)
plt.rcParams['axes.labelsize'] = '20'
plt.rcParams['axes.labelpad']='8'
plt.rcParams['xtick.labelsize'] = 'xx-large'
plt.rcParams['ytick.labelsize'] = 'xx-large'
plt.subplots_adjust(wspace=0.2)
plt.subplot(1, 2, 1)
data.groupby(by='album')['duration'].mean().plot(kind='bar');
plt.subplot(1, 2, 2)
data.groupby(by='album')['duration'].std().plot(kind='bar');
```

# Audio Features

## Energy

```python
plt.rcParams['figure.figsize'] = (12,4)
plt.subplots_adjust(wspace=1.0)
plt.subplot(1, 2, 1)
data.groupby(by='album')['energy'].mean().plot(kind='bar');
plt.subplot(1, 2, 2)
data.groupby(by='album')['energy'].std().plot(kind='bar');
```

```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1, 2, 1)
plt.xlabel('Energy')
plt.title('HybridTheory - Meteora')
data[data.album.isin(['Hybrid', 'Meteora'])].energy.plot.hist(bins=10, ylim=[0,20], range=(0,1));
plt.subplot(1, 2, 2)
plt.xlabel('Energy')
plt.title('Minutes - ThousandSuns')
data[data.album.isin(['Minutes', 'Thousandsuns'])].energy.plot.hist(bins=10, range=(0,1), ylim=[0,20]);
```

```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1, 2, 1)
plt.xlabel('Energy')
plt.title('LivingThings - HuntingParty')
plt.rcParams['figure.figsize'] = (16,4)
data[data.album.isin(['Huntingparty', 'Livingthings'])].energy.plot.hist(bins=10, range=(0,1), ylim=[0,20]);
plt.subplot(1, 2, 2)
plt.xlabel('Energy')
plt.title('OneMoreLight')
data[data.album.isin(['Onemorelight'])].energy.plot.hist(bins=10, range=(0,1), ylim=[0,10]);
```

## Loudness


```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1, 2, 1)
data.groupby(by='album')['loudness'].mean().plot(kind='bar');
plt.subplot(1, 2, 2)
data.groupby(by='album')['loudness'].std().plot(kind='bar');
```

```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1, 2, 1)
plt.xlabel('Loudness')
plt.title('HybridTheory - Meteora')
plt.rcParams['figure.figsize'] = (16,4)
data[data.album.isin(['Hybrid', 'Meteora'])].loudness.plot.hist(bins=10, ylim=[0,10], range=(-8,-2));
plt.subplot(1, 2, 2)
plt.xlabel('Loudness')
plt.title('Minutes - ThousandSuns')
data[data.album.isin(['Minutes', 'Thousandsuns'])].loudness.plot.hist(bins=10, range=(-8,-2), ylim=[0,10]);
```

```python
plt.subplots_adjust(wspace=0.5)
plt.subplot(1, 2, 1)
plt.xlabel('Loudness')
plt.title('LivingThings - HuntingParty')
plt.rcParams['figure.figsize'] = (16,4)
data[data.album.isin(['Huntingparty', 'Livingthings'])].loudness.plot.hist(bins=10, range=(-8,-2), ylim=[0,10]);
plt.subplot(1, 2, 2)
plt.xlabel('Loudness')
plt.title('OneMoreLight')
data[data.album.isin(['Onemorelight'])].loudness.plot.hist(bins=10, range=(-8,-2), ylim=[0,5]);
```
